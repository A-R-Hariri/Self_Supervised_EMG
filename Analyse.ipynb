```python
import sys
import pandas as pd
```

```python
csv_path = "within_cnn_d0.csv"
df = pd.read_csv(csv_path)

# ensure types
df["seed"] = df["seed"].astype(int)
df["subject"] = df["subject"].astype(int)
df["accuracy"] = df["accuracy"].astype(float)

def mean_std(g):
    return g.agg(["mean", "std"]).reset_index()

# ======== Cross-subject aggregated, cross-seed aggregated ========
# per test set and per experiment: mean ± std over ALL rows (subjects+seeds)
a1 = mean_std(df.groupby(["experiment", "test_type"])["accuracy"])
print("\n# Cross-subject + cross-seed (all pooled) | by experiment x test_type")
print(a1.to_string(index=False))

# # ======== Cross-seed aggregated, per subject ========
# # per subject, per exp and per test set: mean ± std over seeds (within subject)
# a2 = mean_std(df.groupby(["subject", "experiment", "test_type"])["accuracy"])
# print("\n# Cross-seed aggregated | by subject x experiment x test_type")
# print(a2.to_string(index=False))

# # ======== Show cross-subject aggregated per seed ========
# a3 = mean_std(df.groupby(["seed", "experiment", "test_type"])["accuracy"])
# print("\n# Cross-subject aggregated | by seed x experiment x test_type")
# print(a3.to_string(index=False))
```
